```python
import pandas as pd
import requests
import json
import pickle
import os

y='2017'
with open(os.path.join('Downloads', 'oral_arguments', 'data', 'cases_metadata', y+'.pkl'), 'rb') as f:
    all_case_metadata = pickle.load(f)
```

```python
output = {}
for a in all_case_metadata:
    try: 
        href = a[2]['oral_argument_audio'][0]['href'] 
    except Exception as e:
        href = None
    output[a[2]['ID']] = href
```

```python
output
```

```python
import requests
from time import sleep

for k, v in output.items():
    if v is not None:
        transcript = requests.get(v).json()
        with open(os.path.join('Downloads', 'oral_arguments', 'data', 'transcripts', y, str(k)+'.json'), 'w') as f:
            json.dump(transcript, f, indent=4)
        sleep(3)
    else:
        print(k, "transcript unavailable.")
```
